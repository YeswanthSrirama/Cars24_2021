```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Cars24.csv")
```

```python
data = data.drop(["Unnamed: 0", "EMI (monthly)"], axis=1)
data
```

```python
data = data[data["Price"]<1500000]
```

```python
cat_cols = ["Car Brand", "Model Year", "Location", "Fuel", "Gear", "Ownership"]
```

```python
data[cat_cols]
```

```python
sns.displot(data=data, x="Price")
```

Data is a little Right skewed as expected

```python
plt.figure(figsize=(24,8))
for i in cat_cols:
    data[i].value_counts().plot(kind='bar')
    plt.show();
```

```python
plt.figure(figsize=(24,200))
data["Model"].value_counts().sort_values().plot(kind='barh')
plt.show();
```

```python
# plt.figure(figsize=(24,8))
for i in cat_cols:
    plt.figure(figsize=(24,8))
    sns.barplot(data=data, x=i, y="Price", hue="Ownership")
    plt.show()
```

## Insights
1. There are few cases where cars with Ownership:4 are expensive than Ownership:2,3. The reasons might be
> Those cars travelled less Distance, 
> Bought and sold within a short time
2. Not all the Car Brands have all ownership types.
3. Automatic cars are expensive compared to Manual
4. Petrol+CNG and Electirc cars must be expensive than Diesel. Since these two types are just evolving there is no availabiity of such data which resulting in Diesel as expensive.
